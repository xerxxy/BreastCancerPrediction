```python
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(40)
no_entries = 60 
x = np.sort(np.random.uniform(0,20,no_entries))
noise = np.random.normal(4,15, no_entries)

y = 1.5* x ** 2 + 2 * x + 5 + noise

def f(w,x):
    return w[0] + w[1] * x + w[2] * x**2

def J(w,x):
    return  1/120*np.sum((f(w,x)-y)**2)

def sub(w):
    return f(w,x) - y

w=[1,2,2]
l_rate = 0.00001

for i in range(1000):
    sw0 = w[0] - l_rate *np.sum(sub(w))*1/60
    sw1 = w[1] - l_rate *np.sum(sub(w)*x)*1/60
    sw2 = w[2] - l_rate *np.sum(sub(w)*(x**2))*1/60
    w[0]=sw0
    w[1]=sw1
    w[2]=sw2


print(w)

test = f(w,x)
plt.plot(x,test,color='r')
plt.scatter(x,y,color='green')
plt.show()

```
