```python
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(30)

no_entries = 60
x = np.random.uniform(0,20, no_entries) # we do this for generating those 60 points 
noise = np.random.normal(4,15, no_entries)
entries = 1.5* x ** 2 + 2 * x + 5 + noise


data = np.array([x, entries])



def f(x,w0,w1,w2):
    return w0 + x * w1 + (x**2) * w2  # this is the polynomial regression model

first_col = [1] * 60
second_col = x
third_col = x**2

A = np.array([first_col, second_col, third_col]).T

B = np.array([entries]).T

ATA = np.matmul(A.T, A)
#print(ATA)
NATA = np.linalg.inv(ATA)
#print(NATA)
W = np.matmul(np.matmul(NATA,A.T), B)
#print(W)


x_sorted = np.sort(x)
res = f(x_sorted,W[0],W[1],W[2])



print(W)

# we will compute W = (A^T * A)^(-1) * A^T * B
plt.scatter(x,entries, color = 'blue')
plt.plot(x_sorted,res, color = 'r')
plt.show()


```
