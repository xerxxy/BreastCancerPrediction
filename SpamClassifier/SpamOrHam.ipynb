```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
```

```python
file_path = r'C:\Users\antoc\OneDrive\Desktop\6M\Month1\MLmodels\nBayes\spamham.csv'

data = pd.read_csv(file_path, usecols=range(2))
data.columns = ['label', 'text']
print(data.head())
print(data.iloc[0,1]) # iloc gets the el at pos [0,1] in the dataframe
```

```python
Xi = data['text'] # features
yi = data['label'] # labels

X_train, X_test, y_train, y_test = train_test_split(Xi,yi,test_size=0.5,random_state=40)# r state getting the same data everytime, we seed the generator
print(len(X_train), len(y_train))
```

```python


class NaiveBayesClassifier:
    
    def __init__(self, X, y):
        self.X = list(X)
        self.y = list(y)
        self.spam_prior = None
        self.ham_prior = None 
        self.vocabulary = None 
        self.bigdocHam = []
        self.bigdocSpam = []
        self.loglikelihoodSpam = {}
        self.loglikelihoodHam = {}
        
        

    def tokenize(self, text):
        # Built this for aesthetics
        return text.split()
    
    def build_bigdoc(self):

  
        for i in range(len(self.X)):
            if self.y[i] == "ham":
                self.bigdocHam.extend(self.X[i].split())
            else:
                self.bigdocSpam.extend(self.X[i].split())

                
    
    def build_vocabulary(self):
        self.vocabulary = set()
    
        for document in self.X: 
            words = document.split()
            self.vocabulary.update(words)
        
        self.vocabulary = sorted(self.vocabulary)
 
    

    
    def train(self):
        num_spam = sum(1 for label in self.y if label == 'spam')

        num_ham = len(y_train) - num_spam

        
        self.spam_prior = num_spam / len(self.y)
        self.ham_prior = num_ham / len(self.y)
        

        
        self.spam_prior = np.log(self.spam_prior)
        self.ham_prior = np.log(self.ham_prior)
        self.build_vocabulary()
        self.build_bigdoc()  

        

        denominatorHam = sum(self.bigdocHam.count(w) for w in self.vocabulary) + len(self.vocabulary)
        denominatorSpam = sum(self.bigdocSpam.count(w) for w in self.vocabulary) + len(self.vocabulary)
        
        for word in self.vocabulary:
            count_ham = self.bigdocHam.count(word) + 1
            count_spam = self.bigdocSpam.count(word) + 1
            self.loglikelihoodHam[word] = np.log(count_ham / denominatorHam)
            self.loglikelihoodSpam[word] = np.log(count_spam / denominatorSpam)
            
    
            
            
                
                
                
            
            
            
            
                 


classifier = NaiveBayesClassifier(X_train, y_train)

classifier.train()
probSpam = classifier.spam_prior
probHam = classifier.ham_prior


test_expression = 'lucky, you have been selected to win dollars 10000, click WIN WIN WIN'
test_tokenized = test_expression.split()
for word in test_tokenized:
    if word in classifier.vocabulary:
        print(f"Word: {word}")
        print(f"Log Likelihood Spam: {classifier.loglikelihoodSpam[word]}")
        print(f"Log Likelihood Ham: {classifier.loglikelihoodHam[word]}")
        probSpam += classifier.loglikelihoodSpam[word]
        probHam += classifier.loglikelihoodHam[word]

if probSpam > probHam:
    print('Spam')
else:
   print('Ham')


# print(len(classifier.bigdocSpam))
# print(len(classifier.bigdocHam))
```
